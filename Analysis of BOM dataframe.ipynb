#### Import libraries


```python
import pandas as pd
import numpy as np
```

#### Set options of displaying data

#### Creating dataframe from excel files

```python
#Project excel
projektowy_df = pd.read_excel(
    r'.\excel files\Materiał.xlsx',
    sheet_name='BOM_POPC2_2.0')
#Delivery report
raport_dostaw_df = pd.read_excel(r'.\excel files\Raport dostaw.xlsx',sheet_name=0)
```

#### Selecting excatly columns from projektowy_df which we want to merge

```python
#Selecting three columns from project excel 
columns_project_df=projektowy_df[['Code', 'Unnamed: 7', 'Quantity','Unnamed: 3']][3:]
#Changed names of them, to make our work easily
columns_project_df=columns_project_df.rename(columns = {'Code': 'Kod', 'Unnamed: 7':'Kod vol.2', 'Unnamed: 3':'Nazwa produktu', 'Quantity':'Ilość'}, inplace = False)
#Print out result
columns_project_df.head(20)
```

#### Selecting values starting: "M" and "3", using the startswith method

```python
#Deleted NaN values from column 'Kod'
columns_project_df=columns_project_df.dropna(subset=['Kod'])
#Selecting values starting: "M" and "3", using the startswith method
columns_project_df=columns_project_df[columns_project_df.Kod.str.startswith('M-', '3')]
#Print out result
columns_project_df.head(20)
```

#### Choosing of project with we want to work 

```python
#Function to selecting from delivery report a directly project
def choosing_a_project(name):
    project = raport_dostaw_df[raport_dostaw_df['NAZWA']==name]
    return project

choosen_project = choosing_a_project('Zagórze')
#Choosing columns
code_column_choosen_project = choosen_project[[ 'Code', 'Delivered q-ty ', 'Delivery date']]
#Renaming columns
code_column_choosen_project=code_column_choosen_project.rename(columns = {'Code': 'Kod z dostawy', 'Delivered q-ty ':'Dostarczona ilość', 'Delivery date':'Data dostawy'}, inplace = False)
code_column_choosen_project
```

#### Merging two dataframe together

```python
#Merging two dataframe with left type of merge
merged_df = columns_project_df.merge(code_column_choosen_project, how='left',left_on='Kod', right_on = 'Kod z dostawy')

#Changing NaN to 0
merged_df=merged_df.fillna(0)

#Added condition to drop every 0 value from dataframe if both Columns got 0 value
merged_df = merged_df[(merged_df['Ilość'] != 0) | (merged_df['Dostarczona ilość'] != 0)]

# Changed an arrangement of columns
merged_df_2=merged_df[['Kod', 'Kod vol.2', 'Kod z dostawy','Nazwa produktu', 'Ilość', 'Dostarczona ilość','Data dostawy']]

#Print out result
merged_df_2

```
